2. Who is the top scorer or the top batsman?
3. Which batsman has the highest score?
4. Who is the best bowler?
5. What are the highest number of wickets?
6. Player with the most Player Of The Match award?
7. After the toss, how many have chosen to bat or to bowl first?
8. First innings score V/S Second innings score
9. First innings wickets V/S Second innings wickets
10. To find if the teams have won the match by runs or wickets?

```python
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset.
df = pd.read_csv("t20-world-cup-22.csv")
df
```

```python
#Question 1 - Which country has won the most matches?
sns.set_style("whitegrid")
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(x='winner',data=df)
```

```python
#Question 2 - Who is the top scorer or the top batsman?
sns.set_style("whitegrid")
plt.figure(figsize=(20,5),dpi=200)
plt.xticks(rotation=45)
sns.countplot(x='top scorer',data=df)
```

```python
#Question 3 - Which batsman has the highest score?
sns.set_style("whitegrid")
plt.figure(figsize=(20,5),dpi=200)
plt.xticks(rotation=45)
sns.barplot(x='top scorer',y='highest score',data=df)
```

```python
#Question 4 - Who is called as the best bowler of the series?
sns.set_style("whitegrid")
plt.figure(figsize=(20,5),dpi=200)
plt.xticks(rotation=70)
sns.countplot(x='best bowler',data=df)
```

```python
#Question 5 - What are the highest number of wickets and by whom?
df['Wickets'] =  pd.to_numeric(df['best bowling figure'].str.split('-').str[0],errors='coerce')
df
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(20,5),dpi=200)
plt.xticks(rotation=75)
sns.barplot(x='best bowler',y='Wickets',data=df)
```

```python
#Question 6 - Player with the most Player Of The Match award?
sns.set_style("whitegrid")
plt.figure(figsize=(20,8),dpi=200)
plt.xticks(rotation=75)
plt.yticks(range(0,3))
sns.countplot(x='player of the match',data=df)
```

```python
df['player of the match'].value_counts()
```

```python
choice_counts = df['toss decision'].value_counts()
choice_counts
```

```python
#Question 7 - After the toss, how many have chosen to bat or to bowl first?
plt.figure(figsize=(12,5),dpi=200)
plt.pie(choice_counts,labels=choice_counts.index,autopct='%1.1f%%', startangle=90,data=df)
plt.axis('equal')
```

```python
#Question 8 - First innings score V/S Second innings score
sns.set_style("whitegrid")
plt.figure(figsize=(20,8),dpi=200)
#sns.scatterplot(x='first innings score', y='second innings score', data=df)

sns.scatterplot(x='venue', y='first innings score', data=df, label='First Innings', color='skyblue',s=150)
sns.scatterplot(x='venue', y='second innings score', data=df, label='Second Innings', color='orange', alpha=0.7,s=150)
```

```python
#Question 8 - First innings score V/S Second innings score
sns.set_style("whitegrid")
plt.figure(figsize=(20,8),dpi=200)
plt.ylim(0,15)
sns.scatterplot(x='venue', y='first innings wickets', data=df, label='First Innings', color='skyblue',s=150)
sns.scatterplot(x='venue', y='second innings wickets', data=df, label='Second Innings', color='orange', alpha=0.7,s=150)
```
